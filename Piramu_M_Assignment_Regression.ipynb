**A.	Explain the main assumptions of Linear Regression in detail.**

The 5 assumptions of Linear Regression are as follows:

1.Linearity:
The relationship between the dependent variable and the independent variables should be linear. This means that the effect of the predictors on the outcome is constant. A linear regression model assumes this form:
y=ùõΩ0+Œ≤1x+œµ
If the relationship is not linear, the model‚Äôs predictions may be inaccurate.

2.Independence of Errors:
The residuals (errors) should be independent of each other. This means that the value of the error term for one observation should not depend on another. If errors are correlated (common in time series data), it can lead to biased estimates and incorrect statistical tests.

3.Homoscedasticity:
The variance of the errors should be constant across all levels of the independent variables. When the variance changes (heteroscedasticity), the standard errors of coefficients become unreliable, affecting confidence intervals and hypothesis testing.

4.No Multicollinearity:
Independent variables should not be highly correlated with each other. High multicollinearity makes it difficult to isolate the individual effect of each variable, leading to unstable and misleading coefficient estimates.

5.Normality of Residuals:
The residuals should follow a normal distribution. This assumption is especially important when performing statistical tests (e.g., t-tests for coefficients), as many of these tests rely on normality for valid results.

**D.	How logistic regression works for multiclass classification. Explain in detail.**

Logistic Regression is primarily used for binary classification, but it can be extended to handle multiclass classification problems using techniques like One-vs-Rest (OvR) and Multinomial Logistic Regression.
Multinomial Logistic Regression (Softmax)

building a machine learning model to predict the type of fruit based on features like color, weight, and texture. The possible classes are:

Apple
Banana
Orange

This is a multiclass classification problem because there are more than two possible outcomes.

1. One-vs-Rest (OvR)
In this method, logistic regression treats each class separately. It builds one model per class:

Model 1: Apple vs. (Banana + Orange)
Model 2: Banana vs. (Apple + Orange)
Model 3: Orange vs. (Apple + Banana)

Each model will output the probability of the fruit being a specific class. For example:

Model 1 predicts 0.7 (70% chance it's an Apple)
Model 2 predicts 0.2 (20% chance it's a Banana)
Model 3 predicts 0.1 (10% chance it's an Orange)

The class with the highest probability (Apple) is chosen as the final prediction.

2: Multinomial Logistic Regression (Softmax)
Instead of training multiple models, this approach uses a single model that considers all three classes at once.

For a new fruit, the model directly predicts the probability of being:

Apple: 70%
Banana: 20%
Orange: 10%

Again, the class with the highest probability (Apple) is selected as the output.

**E.	Explain the performance metrics of logistic regression.**

Logistic regression is a classification algorithm, so its performance is evaluated using classification metrics. The most common metrics include:

Accuracy
Measures the proportion of correct predictions (both true positives and true negatives) out of all predictions.
Best used when the dataset is balanced (i.e., equal number of positive and negative cases).

2. Precision
Precision is the proportion of correct positive predictions out of all predicted positives.
High precision means fewer false positives.
Useful when false positives are costly (e.g., predicting someone has a disease when they don‚Äôt).

3. Recall (Sensitivity or True Positive Rate)
Recall is the proportion of actual positives correctly identified.
High recall means fewer false negatives.
Important when missing a positive case is critical (e.g., fraud detection or cancer diagnosis).

4. F1-Score
The harmonic mean of precision and recall.
Useful when you need a balance between precision and recall, especially in imbalanced datasets.

5. ROC-AUC Score (Receiver Operating Characteristic ‚Äì Area Under Curve)
Measures how well the model separates positive and negative classes across different thresholds.
AUC = 1 is perfect; AUC = 0.5 means no better than random guessing.
Useful for comparing models and handling imbalanced classes.

6. Confusion Matrix
A table showing True Positives (TP), True Negatives (TN), False Positives (FP), and False Negatives (FN).
Helps visualize how well the model is performing across classes.

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score # Changed metrics for regression
from sklearn.preprocessing import StandardScaler # Import StandardScaler
import pandas as pd # Import pandas if not already imported

# Assuming df1 is already loaded from the previous cells

# Separate features (X) and target (y)
X = df1.drop(columns=['Product_id', 'Price']) # Features, excluding Product_id and Price
y = df1['Price'] # Target variable is Price

# Perform feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=X.columns) # Convert scaled data back to DataFrame

# Split the scaled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_scaled, y, test_size=0.2, random_state=42)

models = {
    'Linear': LinearRegression(),
    'Ridge': Ridge(alpha=1),
    'Lasso': Lasso(alpha=0.01),
    'ElasticNet': ElasticNet(alpha=0.01, l1_ratio=0.5)
}

# Evaluate regression models (using appropriate metrics like MSE and R2)
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    # Calculate regression metrics
    mse = mean_squared_error(y_test, preds)
    r2 = r2_score(y_test, preds)

    print(f"{name} Mean Squared Error: {mse:.2f}")
    print(f"{name} R-squared: {r2:.2f}")
    print("-" * 30)

# Removed the extra 'W' at the end
```
